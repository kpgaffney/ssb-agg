```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set()

df = pd.read_csv('results.csv')

df = df.melt(
    id_vars=['Query', 'Trial'],
    var_name='Variant',
    value_name='Time (s)'
)

df['Time (ms)'] = df['Time (s)'] * 1000

plt.figure(figsize=(10, 5))

sns.barplot(data=df, x='Query', y='Time (ms)', hue='Variant')

ax = plt.gca()
for patch in ax.patches:
    ax.annotate(
        round(patch.get_height()),
        (patch.get_x() + 0.5 * patch.get_width(), patch.get_height() + 15),
        alpha=0.8,
        size='small',
        ha='center',
        va='top'
    )
        
plt.tight_layout()
plt.savefig('ssb.pdf')
```
